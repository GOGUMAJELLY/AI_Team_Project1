```python
import pandas as pd
import numpy as np

fukuoka = pd.read_csv("Fukuoka_hotel_data.csv")
osaka = pd.read_csv("Osaka_hotel_data.csv")
fukuoka
```

```python
def engineering(data):
    # month + day = datetime 변경
    data['date'] = pd.to_datetime('2023-' + data['month'].astype(str) + '-' + data['day'].astype(str))
    data.drop(columns=['month', 'day'], inplace=True)
    # price 컬럼 원~ 제거후 int 변경
    data['price'] = data['price'].replace({'[,]' : '', '[원~]': ''}, regex=True).astype('int')
    # 호텔 등급 성급 제거 후 int 변경
    data['hotel_star'] = data['hotel_star'].replace({'[성급]': ''}, regex=True).astype('int')

engineering(fukuoka)
engineering(osaka)
```

```python
fukuoka = fukuoka[['date', 'hotel_name', 'price', 'hotel_star', 'review_score', 'hotel_URL']]
osaka = osaka[['date', 'hotel_name', 'price', 'hotel_star', 'review_score', 'hotel_URL']]
osaka = osaka.replace('평점없음', np.nan)
fukuoka
```

```python
fukuoka.to_csv('./Fukuoka_hotel.csv')
osaka.to_csv('./Osaka_hotel.csv')
```
